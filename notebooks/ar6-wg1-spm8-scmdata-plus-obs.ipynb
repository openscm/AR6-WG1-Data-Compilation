# AR6 WG1 - SPM.8 plus observations

This notebook combines parts of **Figure SPM.8** and **Figure SPM.10** of the IPCC's *Working Group I contribution to the Sixth Assessment Report* ([AR6 WG1](https://www.ipcc.ch/assessment-report/ar6/)). It follows the work by Ed Hawkin's shown at [Climate Lab Book](http://www.climate-lab-book.ac.uk/2021/adding-observations/).

The data supporting the SPM figure is published under a Creative Commons CC-BY license at 
the [Centre for Environmental Data Analyis (CEDA)](https://catalogue.ceda.ac.uk/uuid/ae4f1eb6fce24adcb92ddca1a7838a5c).

This notebook uses a version of that data which was processed for interoperability with the format used by IPCC WG3, the so-called IAMC format.  
The notebook is available under an open-source [BSD-3 License](https://github.com/openscm/AR6-WG1-Data-Compilation/blob/main/LICENSE) in the [openscm/AR6-WG1-Data-Compilation](https://github.com/openscm/AR6-WG1-Data-Compilation) GitHub repository.

The notebook uses the community Python package [scmdata](https://scmdata.readthedocs.io), which is designed to handle simple climate model data but generalises to handle all sorts of timeseries-based data including data in the IAMC format. While it is primarily a data handling package and not a plotting package, scmdata does offer some convenient plotting functions which we use in these notebooks alongside [seaborn](https://seaborn.pydata.org) and native [matplotlib](https://matplotlib.org/) as required.

```python
import matplotlib.pyplot as plt
import pyam
import scmdata
import yaml

import utils
```

```python
plt.rcParams["axes.xmargin"] = 0
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
with open("plotting.yaml") as fh:
    palette = yaml.safe_load(fh)["color"]["scenario"]
    
palette = {k: pyam.plotting.PYAM_COLORS[v] for k, v in palette.items()}
palette["historical"] = "black"
palette
```

```python
input_file_spm8 = utils.DATA_DIR / "processed" / "fig-spm8" / "fig-spm8-timeseries.csv"
input_file_spm8
```

```python
data_spm8 = scmdata.ScmRun(str(input_file_spm8))
data_spm8.head()
```

```python
input_file_spm10 = utils.DATA_DIR / "processed" / "fig-spm10" / "fig-spm10-timeseries.csv"
input_file_spm10
```

```python
data_spm10 = scmdata.ScmRun(str(input_file_spm10))
data_spm10.head()
```

```python
# the output format would work nicely with pyam, openscm prefers
# variables to not contain stats ids (for example)
data_spm8_unmangled = utils.unmangle_stats(data_spm8)
data_spm8_unmangled["quantile"] = data_spm8_unmangled["statistic"].apply(utils.convert_stats_to_quantile)
data_spm8_unmangled.head()
```

```python
# the output format would work nicely with pyam, openscm prefers
# variables to not contain stats ids (for example)
data_spm10_unmangled = utils.unmangle_stats(data_spm10.filter(model="Ch.2 assessment"))
data_spm10_unmangled["quantile"] = data_spm10_unmangled["statistic"].apply(utils.convert_stats_to_quantile)
data_spm10_unmangled.head()
```

```python
plt_years = range(1950, 2100 + 1)
# plt_years = range(1950, 2030 + 1)

pdf = (
    data_spm10_unmangled
    .append(data_spm8_unmangled.filter(variable="Surface Air Temperature Change"))
    .filter(year=plt_years)
)
pdf["source_variable"] = pdf["model"] + " - " + pdf["variable"]

dashes = {
    "Ch.2 assessment - Surface Air Temperature Change": "-",
    "Ch.2 assessment - Surface Air Temperature Change|Anthropogenic": "--",
    "Ch.4 Assessed - Surface Air Temperature Change": "-"
}
ax, lh = pdf.plumeplot(
    hue_var="scenario", hue_label="Scenario", palette=palette, 
    style_var="source_variable", style_label="Source - variable", dashes=dashes,
    pre_calculated=True,
    quantiles_plumes=[((0.05, 0.95), 0.1), ((0.5,), 1.0)],
)

(
    pdf
    .filter(source_variable="Ch.2 assessment - Surface Air Temperature Change")
    .lineplot(linewidth=5, color="black", ax=ax, legend=False)
)
```
