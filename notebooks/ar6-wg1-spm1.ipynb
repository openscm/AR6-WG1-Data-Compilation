# AR6 WG1 - SPM.1

This notebook replicates AR6 WG1 Figure SPM.1.

```python
import matplotlib.pyplot as plt
import scmdata

import utils
```

```python
plt.rcParams["axes.xmargin"] = 0
```

```python
input_file = utils.DATA_DIR / "processed" / "fig-spm1" / "fig-spm1-timeseries.csv"
input_file
```

```python
data = scmdata.ScmRun(str(input_file))
data.head()
```

```python
# the output format would work nicely with pyam, openscm prefers
# variables to not contain stats ids (for example)
data_unmangled = utils.unmangle_stats(data)
data_unmangled["quantile"] = data_unmangled["statistic"].apply(utils.convert_stats_to_quantile)
data_unmangled.head()
```

```python
time_axis = "year"

palette = {
    "direct observations": "black",
    "paleoclimate archives": "tab:blue",
    "hist-nat": "tab:green",
    "hist-ssp245": "tab:orange",
    "historical": "black",
}
hue = "model"
hue_label = "Source"
plume_alpha = 0.2

fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(16, 8))

axes[0], plume_handles = data_unmangled.filter(model=["paleoclimate archives", "direct observations"]).plumeplot(
    hue_var=hue, hue_label=hue_label, palette=palette, pre_calculated=True, ax=axes[0], time_axis=time_axis, 
    quantiles_plumes=[((0.05, 0.95), 0.2), ((0.5,), 1.0)],
)

axes[0].axvspan(1850, 2020, color="gray", alpha=plume_alpha)

hue = "scenario"
hue_label = "Scenario"
style="model"
style_label="Source"
dashes = {
    "HadCRUT4.6": "-",
    "CMIP6 multi-model ensemble": "-",
}
axes[1], plume_handles = data_unmangled.filter(model=["HadCRUT4.6", "CMIP6 multi-model ensemble"]).plumeplot(
    hue_var="scenario", hue_label=hue_label, palette=palette, 
    style_var=style, style_label=style_label, dashes=dashes,
    pre_calculated=True, ax=axes[1], time_axis=time_axis, 
    quantiles_plumes=[((0.05, 0.95), 0.2), ((0.5,), 1.0)],
)

axes[0].set_ylim([-1, 2])
axes[0].set_ylabel(
    f"{data_unmangled.get_unique_meta('variable', True)} "
    f"(rel. to "
    f"{data_unmangled.get_unique_meta('reference_period_start_year', True)}"
    f"-{data_unmangled.get_unique_meta('reference_period_start_year', True)}, "
    f"{data_unmangled.get_unique_meta('unit', True)})"
)
for ax in axes:
    ax.grid(axis="y")
```
